```python
import matplotlib.pyplot as plt
from pydicom import dcmread
from pydicom.data import get_testdata_file

fpath = get_testdata_file('tumors/Calc-Test_P_00038_LEFT_CC/08-29-2017-DDSM-96009/1.000000-full mammogram images-63992/1-1.dcm')
ds = dcmread(fpath)
```

```python
display_name = ds.PatientName.family_name + ", " + ds.PatientName.given_name
print(f"Patient: {display_name}")
print(f"ID: {ds.PatientID}")
```
